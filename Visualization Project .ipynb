### Business Question 2:

any kind of correlation between all the columns present in the dataset.

### Business Question 3:
#### Which city has the most sales overall?

Here im using the ax.pie() which is a built-in function used for visualization 
and im visualizing the  which show the most sales according to the city.

```python
fig, ax = plt.subplots(figsize =(10, 7))
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
explode = (0.2, 0.1, 0.0)

x = arty["City"].value_counts()

wedges, texts, autotexts = ax.pie(x, labels = ['Yangon', 'Mandalay', 'Naypyitaw'], autopct = "%.2f%%", wedgeprops = wp, shadow = True, explode = explode);
ax.set_title("Most sales according to city:",fontsize = 15);
ax.legend(wedges, arty.City,
          title ="City",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
```

The above illustration shows that 'Yangon'sity has the most sales out of the 
other cities.

### Business Question 4:
#### According to the given dataset in which month most products are sold?

Here im using the ax,pie() which is a built-in function used for visualization and im visualizing the relation betwwen the colums "Total" and "Month" 
which shows the most sold products according to months

```python
fig, ax = plt.subplots(figsize = (10, 5))
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
explode = (0.0, 0.2, 0.1)

wedges, texts, autotexts = ax.pie(arty.groupby(["Month"])["Total"].agg(["sum"])["sum"] , labels = arty.groupby(["Month"])["Total"].agg(["sum"]).index, autopct = "%.2f%%", wedgeprops = wp, shadow = True, explode = explode);
ax.set_title("Most sold products according to months:",fontsize = 15);
ax.legend(wedges, arty.Month,
          title = "Months",
          loc = "right",
          bbox_to_anchor = (1, 0, 0.5, 1))
```

According to the above pie chart in the month of 'January' most products are sold.

### Business Question 5:
#### Which branch sells more products in each month?

Here im using the sns.barplot() which is a built-in function used for visualization and im visualizing the relation betwwen the colums "Total" and "Branch" 
which shows the sale of individual branch in each month

```python
#cp = {"January": "blue", "March": "green", "February": "black"}
sns.set(rc = {'axes.facecolor':'lightblue', 'figure.facecolor':'lightblue'})

plt.figure(figsize = (16,10))
sns.barplot(x = "Total", y = "Branch", data = arty, 
            hue = "Month", edgecolor = 'red')
plt.title("Individual branch sale in each month",fontsize = 20)
```

### Business Question 6:
#### What is the trend of sales of all the branches in three months

Here im using the sns.relplot and linechart which is a built-in function used for visualizing frequency and im visualizing the relation betwwen the colums "Total" and "Date" 
which shows the frequenct of sales on monthly basis

```python
plt.figure(figsize = (15,10))
sns.set(rc = {'axes.facecolor':'lightblue', 'figure.facecolor':'lightblue'})

sns.relplot(x = "Date", y = "Total", data = arty, kind = "line" ,color = 'red', estimator = sum, row = "Branch")
plt.xticks(rotation = 90);
plt.title("Sale Trend of the Branches over the months",fontsize=15)
```
